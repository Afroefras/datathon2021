```python
import pandas as pd
import os
import datetime as dt
```

```python
csv_file = '/Descargas/datathon/datos/CM02_2/pernoctaciones_viajeros_entradas.csv'
```

```python
df = pd.read_csv(csv_file)
```

```python
o_tot = df[(df['origen'] == 'TOTAL')].index
df.drop(o_tot, inplace = True)
```

```python
v_alo = df[(df['indicador'] == "Viajeros alojados")].index
df.drop(v_alo , inplace=True)
```

```python
anios = ["2009","2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]
for x in anios:
    sn = df[(df['periodo'] == x)].index
    df.drop(sn , inplace=True)
```

```python
df["Año"] = df["periodo"].str[:4]
df["Mes"] = df["periodo"].str[-2:]
```

```python
zip_file = '/Descargas/datathon/Catalogo de zipcodes.csv'
zp = pd.read_csv(zip_file)
```

```python
zp
```

```python
df2 = df.merge(zp, left_on = "municipio", right_on = "Municipio")
```

```python
df2[pd.to_numeric(df2["valor"], errors='coerce')
```

```python
df2 = df2[df2.valor != ".."]
```

```python
df2["valor"] = df2["valor"].astype(int)
```

```python
pv = pd.pivot_table(df2, values="valor", index = 'C¢digo INE', columns = 'indicador')
```

```python
pv["NochesPorPersona"] = round(pv["Pernoctaciones"]/pv["Viajeros entrados"])
```
