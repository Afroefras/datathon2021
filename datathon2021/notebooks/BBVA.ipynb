## Parameters

```python
BASE_DIR = '/Users/efraflores/Desktop/EF/Contests/Datathon_202109/data/datos'
CP_FILE = 'Canarias_CP.csv'
FOLDER_NAME = 'CM01_3'
```

## Import

```python
from mariachis.models import Bbva

bbva = Bbva(BASE_DIR, CP_FILE, FOLDER_NAME)
print(bbva)
```

## Transform

```python
df, pipe_obj = bbva.full_pipeline_bbva(comunity_level=True, merge_zipcode=True, cluster_kwargs={'n_clusters':4, 'kmeans':False}, polygon_kwargs={'dissolve_by':'comunity_code'}, index=['comunity_code'], values=['avg_amount','cards'], aggfunc=['count','sum','mean','median','min','max'], fill_value=0)
df.head(2)
```

## Profiles

```python
from mariachis.utils import profiles

prof = profiles(df[['cluster','province','comunity']].join(df.filter(regex='sum|mean|min|max')))

for var in prof.values():
    display(var.fillna(0).style.format("{:.0f}").background_gradient('Blues'))
```

## Export

```python
bbva.export_result(df, 'Finished_BBVA.csv')
```
