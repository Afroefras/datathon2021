## Import

```python
from mariachis.models import Pubs
        
pub = Pubs(BASE_DIR, FOLDER_NAME, FILE_NAME, CP_FILE)
print(pub)
```

## Transform

```python
df, pipe_obj = pub.full_pipeline_pubs(cluster_kwargs={'n_clusters':6, 'kmeans':True}, index='comunity_code', columns='Comercio', aggfunc={'Valor':['count','sum','mean']}, fill_value=0)
df.sample(2)
```

## Profiles

```python
from mariachis.utils import profiles

prof = profiles(df)

for var in prof.values():
    display(var.fillna(0).style.format("{:.0f}").background_gradient('Blues'))
```
