## Parameters

```python
BASE_DIR = '/Users/efraflores/Desktop/EF/Contests/Datathon_202109/data/datos'
CP_FILE = 'Canarias_CP.csv'
FOLDER_NAME = 'CM05'
FROM_YEAR = 2018
TO_YEAR = 2021
```

## Import

```python
from mariachis.models import Weather

we = Weather(BASE_DIR, CP_FILE, FOLDER_NAME, FROM_YEAR, TO_YEAR)
print(we)
```

## Transform

```python
df, pipe_obj= we.full_pipeline_weather(cluster_kwargs={'n_clusters':6, 'kmeans':True}, index=['comunity_code'], columns=['indicador','mes'], aggfunc={'valor':['count','mean','median','min','max']}, fill_value=0)
print(len(df))
df.sample()
```

## Profiles

```python
from mariachis.utils import profiles

prof = profiles(df[['cluster']].join(df.filter(like='mean')))

for var in prof.values():
    display(var.fillna(0).style.format("{:.2f}").background_gradient('Blues'))
```

## Export

```python
we.export_result(df, 'Finished_weather.csv')
```
